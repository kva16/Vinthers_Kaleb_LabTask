### Part 1: Data Exploration ###
1. Print the shape and column names of the dataset.

2. Display basic information and summary statistics about the dataset.

3. Show the first 10 rows.

```python
#Shape and column names of dataset
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

#Basic info and summary statistics about the dataset
print(df.info())
print(df.describe())

#Show the first 10 rows
print(df.head(10))
```

### Part 2: Indexing and Selection ###
1. Select the names column.

2. Select multiple columns of your choice.

3. Select a single row from the dataset.

4. Select a rectangular section (subset of rows and columns).

```python
#Select the names column
print(df["Name"])

#Select multiple columns
print(df[["Name", "Age", "Fare"]])

#Select a single row from the dataset
print(df.loc[0])

#Select a rectangular section
print(df.loc[0:4, ["Name", "Pclass", "Age", "Fare"]])
```

### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

```python
#Show only rows where the age is above 60
print(df[df["Age"] > 60])

#Rows where the fare is above 100 and the passenger was in 1st class
print(df[(df["Fare"] > 100) & (df["Pclass"] == 1)])

#Identify and isolate rows with missing values in the age column
print(df[df["Age"].isna()])
```

### Part 4: Reindexing and Dropping ###

1. Reverse the row order of the dataset.

2. Remove one column from the dataset.

3. Remove all rows with missing values in the age column.

```python
#Reverse the row order of the dataset
reversed_df = df.iloc[::-1] #python slicing technique to go through the rows
print(reversed_df)

#Remove one column from the dataset
df_dropped = df.drop(columns=["Cabin"]) #if drop is true dont add a new column
print(df_dropped.head())

#Remove all rows with missing values in the age column
df_no_missing_age = df.dropna(subset=["Age"]) # Limits only to the age column
print(df_no_missing_age)
```

### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

```python
#Create a new column that converts fares from USD to Euros
df["Fare_Euro"] = df["Fare"] * 0.92
print(df[["Fare", "Fare_Euro"]].head())

#Subtract the average age from every age value
average_age = df["Age"].mean()
df["Age_Difference"] = df["Age"] - average_age
print(df[["Age", "Age_Difference"]].head())

#Subtract a custom array from two selected columns in the dataset
import numpy as np

custom_array = np.array([1, 10])
result = df[["Age", "Fare"]] - custom_array
print(result.head())
```

### Part 6: Applying Functions ###

1. Create a new column with the number of characters in each name.

2. Calculate the difference between fare and age for each row.

3. Round all numerical values in the dataset to one decimal place.



```python
#Create a new column with the number of characters in each name
df["Name_Length"] = df["Name"].apply(len)
print(df[["Name", "Name_Length"]].head())

#Calculate the difference between fare and age for each row
df["Fare_Age_Difference"] = df["Fare"] - df["Age"]
print(df[["Fare", "Age", "Fare_Age_Difference"]].head())

#Round all numerical values in the dataset to one decimal place
rounded_df = df.round(1)
print(rounded_df.head())
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
#Convert categorical values in the gender column to numeric codes
df["Sex_Code"] = df["Sex"].map({"male": 0, "female": 1})
print(df[["Sex", "Sex_Code"]].head())

#Create a new column that categorizes age into groups
def age_group(age):
    if pd.isna(age):
        return "N/A"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

print(df[["Age", "Age_Group"]].head(20))
```

```python
import urllib.request
import csv
import io

#Task Part 2: Select the Name Column

URL = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"

with urllib.request.urlopen(URL) as response:
    raw = response.read().decode("utf-8")

reader = csv.DictReader(io.StringIO(raw))
data = [row for row in reader]

names = [row["Name"] for row in data]
print(names[:5])
```

```python
#Part 3 Rows that have ages higher than 60

old_passengers= [
    row for row in data
    if row["Age"] != "" and float(row["Age"]) > 60
]

for row in old_passengers:
    print(f"{row['Name']}, Age: {row['Age']}")
```

```python
#Part 5 Convert fare from US currency to Euros
EUR_RATE = 0.92

fares_eur= [
    {"Name": row["Name"], "Fare_USD": float(row["Fare"]), "Fare_EUR": round(float(row["Fare"]) * EUR_RATE, 2)} #uses float and rounding to convert currencies
    for row in data
    if row["Fare"] != ""
]

for row in fares_eur:
    print(f"{row['Name']}: ${row['Fare_USD']} → €{row['Fare_EUR']}")
```

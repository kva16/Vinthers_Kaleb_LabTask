```python
import altair as alt
import pandas as pd


df = pd.read_csv('music_streams.csv')


yearly_trend = df.groupby('year').agg({
    'streams_millions': 'sum',
    'playlist_adds': 'sum'
}).reset_index()

# dictionary
colab_safe_data = alt.Data(values=yearly_trend.to_dict(orient='records'))

# line chart
digital_step_chart = alt.Chart(colab_safe_data).mark_line(
    color='#1DB954',         # brihgt green colour
    strokeWidth=4,
    interpolate='step-after',
    point={'color': '#1DB954', 'size': 80}
).encode(
    x=alt.X('year:O', title='Year', axis=alt.Axis(labelAngle=0, grid=False)),
    y=alt.Y('streams_millions:Q', title='Total Streams in Millions'),
    tooltip=[
        alt.Tooltip('year:O', title='Year'),
        alt.Tooltip('streams_millions:Q', title='Total Streams', format='.1f'),
        alt.Tooltip('playlist_adds:Q', title='Total Playlist Additions', format=',')
    ]
).properties(
    width=550,
    height=300,
    title=alt.TitleParams(
        text="Growth Tracking for Streams and Adds",
        anchor="start"
    )
).configure_view(
    strokeWidth=0 # thinner background
)

# show chart
digital_step_chart
```
